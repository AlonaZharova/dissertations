```python
import argparse
import random
import pdb

features = ['int.rate', 'orig.upb', 'fico.score', 'dti.r', 'ltv.r', 'bal.repaid', 't.act.12m',
       't.del.30d.12m', 't.del.60d.12m', 'hpi.st.d.t.o', 'hpi.zip.o',
       'hpi.zip.d.t.o', 'ppi.c.FRMA', 'TB10Y.d.t.o', 'FRMA30Y.d.t.o',
       'ppi.o.FRMA', 'equity.est', 'hpi.st.log12m', 'hpi.r.st.us',
       'hpi.r.zip.st', 'st.unemp.r12m', 'st.unemp.r3m', 'TB10Y.r12m',
       'T10Y3MM', 'T10Y3MM.r12m']

# feat_dict: featname: set()---distinct feat values
feat_dict = {}


# (max_value +1) +1
feat_size = { 
    # int.rate: 10.875 * 14.71 (distinct 89) {the distance from 4 tp 10 is almost 6, dividing 89 on 6, we get almost 14.71}
    # lower limit 4.75 * 14.71 almost 70, higher limit 10.875 * 14.71 is amost 160
    "int.rate": [78, 174 + 1],#  multi_factor:16.522
    
    "orig.upb": [22000, 465001 +1],
    "fico.score": [300, 821 + 1],
    "dti.r": [1, 66 + 1],
    "ltv.r": [8, 101 + 1], 
    
    "bal.repaid": [-114, 9034 + 1],#  multi_factor:90.333
    
    "t.act.12m": [0, 13 + 1],
    "t.del.30d.12m": [0, 11 + 1],
    "t.del.60d.12m": [0, 13 + 1],
    
    "hpi.st.d.t.o":[-710, 8145 + 1],#  multi_factor:64.37
    "hpi.zip.o":[1363, 5370 + 1],#  multi_factor:15.701
    "hpi.zip.d.t.o":[-205, 4045 + 1],#  multi_factor:19.612
    "ppi.c.FRMA":[-233, 567 + 1],#  multi_factor:122.171
    "TB10Y.d.t.o":[-265, 143 + 1],#  multi_factor:79.648
    "FRMA30Y.d.t.o":[-279, 141 + 1],#  multi_factor:82.643
    "ppi.o.FRMA":[-193, 346 + 1],#  multi_factor:127.943

    "equity.est":[520, 1413749 + 1],
    
    "hpi.st.log12m":[-360, 2245 + 1],#  multi_factor:6943.341
    "hpi.r.st.us":[2206, 4784 + 1],#  multi_factor:3370.954
    "hpi.r.zip.st":[3775, 11639 + 1],#  multi_factor:5104.079
    "st.unemp.r12m":[-29, 23 + 1],#  multi_factor:4.146
    "st.unemp.r3m":[-13, 13 + 1],#  multi_factor:2.066
    "TB10Y.r12m":[-34, 42 + 1],#  multi_factor:21.186
    "T10Y3MM":[-11, 74 + 1],#  multi_factor:19.858
    "T10Y3MM.r12m":[-33, 54 + 1],#  multi_factor:13.522
}

# feat_index
feat_index = {}

# build feat_index
def build_feat_index(feat_index_f):
    max_index = 0
    feat_index_lines = []
    #pdb.set_trace()
    # truncate
    feat_index["0:truncate"] = max_index
    feat_index_lines.append("0:truncate\t{}\n".format(max_index))
    max_index += 1

    for i in range(len(features)):
        feat_num = i + 1
        # other
        feat_index["{}:other".format(feat_num)] = max_index
        feat_index_lines.append("{}:other\t{}\n".format(feat_num, max_index))
        max_index += 1
        # normal values
        low = feat_size[features[i]][0]
        high = feat_size[features[i]][1]
        for value in range(low, high):
            feat_index["{}:{}".format(feat_num, value)] = max_index
            feat_index_lines.append("{}:{}\t{}\n".format(feat_num, value, max_index))
            max_index += 1

    # write feat_index file:
    with open(feat_index_f, 'w') as f:
        f.writelines(feat_index_lines)

    return feat_index, feat_index_lines
 
def get_feat_val(feat_name, str_val):
    if feat_name == "int.rate":
        return int(float(str_val) * 16.522)
    elif feat_name == "bal.repaid":
        return int(float(str_val) * 90.333)
    elif feat_name == "hpi.st.d.t.o":
        return int(float(str_val) * 64.37)
    elif feat_name == "hpi.zip.o":
        return int(float(str_val) * 15.701)
    elif feat_name == "hpi.zip.d.t.o":
        return int(float(str_val) * 19.612)
    elif feat_name == "ppi.c.FRMA":
        return int(float(str_val) * 122.171) 
    if feat_name == "TB10Y.d.t.o":
        return int(float(str_val) * 79.648)    
    elif feat_name == "FRMA30Y.d.t.o":
        return int(float(str_val) * 82.643)    
    elif feat_name == "ppi.o.FRMA":
        return int(float(str_val) * 127.943)    
    elif feat_name == "hpi.st.log12m":
        return int(float(str_val) * 6943.341)    
    elif feat_name == "hpi.r.st.us":
        return int(float(str_val) * 3370.954)    
    elif feat_name == "hpi.r.zip.st":
        return int(float(str_val) * 5104.079)    
    if feat_name == "st.unemp.r12m":
        return int(float(str_val) * 4.146)   
    elif feat_name == "st.unemp.r3m":
        return int(float(str_val) * 2.066)    
    elif feat_name == "TB10Y.r12m":
        return int(float(str_val) * 21.186)
    elif feat_name == "T10Y3MM":
        return int(float(str_val) * 19.858)
    elif feat_name == "T10Y3MM.r12m":
        return int(float(str_val) * 13.522)
    else:
        return  int(float(str_val))

# build yzbx data
# y is always 0 
# z is the true time if e=1 and the randomly predicted time (bigger than the ture time) if e=0
# b is the true time if e=0, and the randomly predicted time (bigger than the ture time) if e=1
# x is the encoded features 
def build_yzbx_data(ft_ind,raw_train, raw_test, yzbx_train, yzbx_test):
    fnames_i = [raw_train, raw_test] # it will only inculde the db names as strings 
    fnames_o = [yzbx_train, yzbx_test] # the destination directories of the resulted files as strings 
    #pdb.set_trace()
    for j in range(len(fnames_i)): # len(fnames_i) wiill be 2
        yzbx_list = []
        with open(fnames_i[j], 'r') as f: # openning the first file, which the train file in the first loop 
            lines = f.readlines() # reading all the lines
            for line in lines:# getting each line after the other 
                # the first line would be '-7,113,29087,26.658065492,9.2,146.45,2.7159028845999997,8.3,0,0,1,0,45000,715,69.4,9.2,87.03,24,1\n'
                x_items = []
                # truncate
                x_items.append("0:1") # append the first value to x_itmes
                line_items = line.split(',') # split the line to items saperated by comma
                # line_items will contain: ['-7', '113', '29087', '26.658065492', '9.2', '146.45', '2.7159028845999997', '8.3', '0', '0', '1', '0', '45000', '715', '69.4', '9.2', '87.03', '24', '1\n']
                for i in range(len(line_items) - 2): # again -2 to exclude the time and lable, because lable and time are in the last position in the dataset
                    # in this for we go through each item in line_items individually 
                    feat_num = i + 1
                    feat_name = features[i] # the first feature is "orig_time"
                    feat_val = get_feat_val(feat_name, line_items[i]) # return integer value, line_items[0] = '-7' "orig_time"
                                                                      # if the feature is LTV_time, then we multiply the true value 26.658065492
                                                                      # with 74.53 yielding feat_val = 1986
                    #"orig_time" feat_val = -7
                    key = "{}:{}".format(feat_num, feat_val) #contain feature number and the real value of the feature 
                    #key = {'1:-7'}
                    if key not in ft_ind.keys(): # if the key deosn't exsit in the feature index file
                        key = "{}:other".format(feat_num)    
                    x_items.append("{}:1".format(ft_ind[key])) # append '35:1" to x_items, where 35 the value assigned to the key {'1:-7'} and 1 is the feature position
                                                               # '1002307:1'LTV_time
                # get b and z
                # t from hour to day
                t = int(float(line_items[-2])) # get the time which is the column before the last, hence -2, 
                                                         #time is 30 here, divide by 24 equal 1.25, convert to int then t is 1, add 1 then the result is 2
                e = int(float(line_items[-1]))
                if e == 1:# if event happens, then we know z the true time and we take a rendom number for 
                    z = t # 24
                    b = random.randint(z + 1, 2 * (z + 1)) # A random integer in range [start, end] including the end points. 
                                                            # randint(24 + 1, 2 * (24 + 1)) = randint(25, 50) = 39
                else: # if cens
                    b = t
                    z = random.randint(b + 1, 2 * (b + 1))
                # dummy y
                y = 0
                yzbx = ' '.join([str(y), str(z), str(b)] + x_items)
                yzbx_list.append(yzbx + '\n')

        with open(fnames_o[j], 'w') as f:
            f.writelines(yzbx_list)

      


# calculate features' statistical: max value, min value, mean value, and different value counts
def feat_stat(train, test):
    for feat in features:
        feat_dict[feat] = set() # we set a dictionary with the features names to add values to it in the for loops below
    #pdb.set_trace()
    # train
    with open(train, 'r') as f:
        lines = f.readlines()
        for line in lines:
            items = line.split(',')
            for i in range(len(items) - 2): # here -2 because we don't want to include the time and the label, which are located in the last two indexes
                feat_name = features[i]
                feat_dict[feat_name].add(float(items[i]))
    # test
    with open(test, 'r') as f:
        lines = f.readlines()
        for line in lines:
            items = line.split(',')
            for i in range(len(items) - 2):
                feat_name = features[i]
                feat_dict[feat_name].add(float(items[i]))
    
    # print result
    for feat in features:
        stats = feat_dict[feat]
        print("{} result:".format(feat))
        print("{0}, {1}, {2}".format(max(stats), min(stats), len(stats)))
```
